average sentence length (in words)

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
import pandas as pd

nltk.download('punkt_tab')
nltk.download('punkt')

def avg_sentence_len(text):
    if not text.strip():
        return 0

    text = text.replace("...", " <ELLIPSIS> ")

    sentences = sent_tokenize(text)

    # Восстанавливаем многоточия
    sentences = [s.replace("<ELLIPSIS>", "...") for s in sentences]

    words = word_tokenize(text)

    if len(sentences) == 0:
        return 0

    return len(words) / len(sentences)

df['avg_sent_len'] = df['transcript'].apply(avg_sentence_len)

print(df)
```

Gunning Fog index

```python
import re

def count_syllables(word):
    word = word.lower()
    vowels = "аеёиоуыэюя"
    return sum(1 for char in word if char in vowels)

def count_complex_words(words):
    complex_words = [word for word in words if count_syllables(word) >= 3]
    return len(complex_words)

def calculate_gunning_fog_index(text):
    sentences = re.split(r'[.!?…]', text)
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]

    words = re.findall(r'\b[А-Яа-яЁё\-]+\b', text)

    total_words = len(words)
    total_sentences = len(sentences)

    complex_words = count_complex_words(words)

    if total_sentences == 0 or total_words == 0:
        return 0

    gfi = 0.4 * ((total_words / total_sentences) + 100 * (complex_words / total_words))
    return round(gfi, 2)
```

```python
indx = []
for element in df['transcript']:
  hci = calculate_gunning_fog_index(str(element))
  indx.append(round(hci, 2))
indx
```

```python
df['gunning_fog_index'] = indx
df
```

type-token ratio (TTR)

```python
import re


def calculate_ttr(text):
    words = re.findall(r'\b\w+\b', text.lower())

    total_words = len(words)
    unique_words = len(set(words))

    if total_words == 0:
        return 0
    ttr = unique_words / total_words

    return ttr
```

```python
res_ttr = []
for element in df['transcript']:
  ttr = calculate_ttr(element)
  res_ttr.append(round(ttr, 2))
res_ttr
```

```python
df['ttr'] = res_ttr
df
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

def calculate_subjectivity_coefficient(text):
    words = word_tokenize(text.lower())

    tagged_words = pos_tag(words)

    subject_tags = {'NN', 'NNS', 'NNP', 'NNPS',
                    'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ',
                    'JJ', 'JJR', 'JJS',
                    'RB', 'RBR', 'RBS'}


    subject_words = [word for word, tag in tagged_words if tag in subject_tags]

    stop_words = set(stopwords.words('russian'))
    meaningful_words = [word for word in words if word.isalpha() and word not in stop_words]

    if len(meaningful_words) == 0:
        return 0
    subjectivity_coefficient = len(subject_words) / len(meaningful_words)

    return subjectivity_coefficient
```

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def calculate_quality_coefficient(text):

    words = word_tokenize(text.lower())

    tagged_words = pos_tag(words)

    adjective_adverb_tags = {'JJ', 'JJR', 'JJS', 'RB', 'RBR', 'RBS'}
    noun_verb_tags = {'NN', 'NNS', 'NNP', 'NNPS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'}

    adjective_adverb_count = sum(1 for word, tag in tagged_words if tag in adjective_adverb_tags)

    noun_verb_count = sum(1 for word, tag in tagged_words if tag in noun_verb_tags)

    if noun_verb_count == 0:
        return float('inf') if adjective_adverb_count > 0 else 0
    quality_coefficient = adjective_adverb_count / noun_verb_count

    return quality_coefficient
```

```python
res_qu = []
for element in df['transcript']:
  coefficient = calculate_quality_coefficient(element)
  res_qu.append(round(coefficient, 2))
res_qu
```

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def calculate_dynamism_coefficient(text):
    words = word_tokenize(text.lower())

    tagged_words = pos_tag(words)

    verb_tags = {'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'}
    noun_adjective_pronoun_tags = {'NN', 'NNS', 'NNP', 'NNPS', 'JJ', 'JJR', 'JJS', 'PRP', 'PRP$', 'WP', 'WP$'}

    verb_count = sum(1 for word, tag in tagged_words if tag in verb_tags)

    noun_adjective_pronoun_count = sum(1 for word, tag in tagged_words if tag in noun_adjective_pronoun_tags)

    if noun_adjective_pronoun_count == 0:
        return float('inf') if verb_count > 0 else 0
    dynamism_coefficient = verb_count / noun_adjective_pronoun_count

    return dynamism_coefficient
```

```python
res_din = []
for element in df['transcript']:
  coefficient = calculate_dynamism_coefficient(element)
  res_din.append(round(coefficient, 2))
res_din
```

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import sent_tokenize, word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def calculate_cohesion_coefficient(text):
    sentences = sent_tokenize(text)

    words = word_tokenize(text.lower())

    tagged_words = pos_tag(words)

    preposition_conjunction_tags = {'IN', 'CC'}

    preposition_conjunction_count = sum(1 for word, tag in tagged_words if tag in preposition_conjunction_tags)

    total_sentences = len(sentences)

    if total_sentences == 0:
        return 0
    cohesion_coefficient = preposition_conjunction_count / total_sentences

    return cohesion_coefficient
```

```python
import pandas as pd
from razdel import sentenize, tokenize
from nltk.corpus import stopwords
from sentence_transformers import SentenceTransformer, util
import numpy as np

model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')

import nltk
nltk.download('stopwords')
russian_stopwords = set(stopwords.words('russian'))
```

```python
df = pd.read_excel("norm.xlsx")
text_column = 'transcript'

def calculate_cohesion(text):
    tokens = [t.text.lower() for t in tokenize(text) if t.text.lower() not in russian_stopwords and t.text.isalpha()]
    if not tokens:
        return 0
    unique_tokens = set(tokens)
    lexical_repetition = 1 - (len(unique_tokens) / len(tokens))
    return lexical_repetition

def calculate_coherence(text):
    sentences = [s.text for s in sentenize(text)]
    if len(sentences) < 2:
        return 1
    embeddings = model.encode(sentences)
    similarities = [util.cos_sim(embeddings[i], embeddings[i+1]).item() for i in range(len(embeddings)-1)]
    return np.mean(similarities)

# Применение к каждому тексту
df['cohesion'] = df[text_column].apply(calculate_cohesion)
df['coherence'] = df[text_column].apply(calculate_coherence)

# Сохранение
df.to_excel("texts_with_cohesion_coherence.xlsx", index=False)
```

```python
import pandas as pd
from natasha import Segmenter, Doc
from pymorphy2 import MorphAnalyzer

df = pd.read_excel('norm.xlsx')
text_column = 'transcript'

segmenter = Segmenter()
morph = MorphAnalyzer()

def count_verb_tenses(text):
    if pd.isna(text) or not isinstance(text, str):
        return pd.Series({'past': 0, 'present': 0, 'future': 0})

    doc = Doc(text)
    doc.segment(segmenter)

    past = 0
    present = 0
    future = 0

    for token in doc.tokens:
        word = token.text
        parse = morph.parse(word)[0]
        if 'VERB' in parse.tag or 'INFN' in parse.tag:
            tense = parse.tag.tense
            if tense == 'past':
                past += 1
            elif tense == 'pres':
                present += 1
            elif tense == 'futr':
                future += 1

    total = past + present + future
    if total == 0:
        return pd.Series({'past': 0, 'present': 0, 'future': 0})

    return pd.Series({
        'past': round(past / total, 3),
        'present': round(present / total, 3),
        'future': round(future / total, 3)
    })

df[['past', 'present', 'future']] = df[text_column].apply(count_verb_tenses)

df.to_excel('verb_tenses_output.xlsx', index=False)
```

```python
import pandas as pd
import spacy

nlp = spacy.load("ru_core_news_sm")


files = {
    'norm': 'norm.xlsx',
    'schizophrenia': 'group_sz.xlsx',
    'depression': 'group_dep.xlsx'
}

pos_tags = ['NOUN', 'VERB', 'ADJ', 'ADV', 'PRON']

def calculate_pos_ratios(text):
    doc = nlp(text)
    total_tokens = len([token for token in doc if not token.is_punct and not token.is_space])
    if total_tokens == 0:
        return pd.Series([0]*len(pos_tags), index=[f'{pos}_ratio' for pos in pos_tags])

    pos_counts = {pos: 0 for pos in pos_tags}

    for token in doc:
        if token.pos_ in pos_counts and not token.is_punct and not token.is_space:
            pos_counts[token.pos_] += 1

    return pd.Series({f'{pos}_ratio': pos_counts[pos] / total_tokens for pos in pos_counts})


medians = {}


for label, path in files.items():
    df = pd.read_excel(path)
    df = df.dropna(subset=['transcript'])

    ratios = df['transcript'].apply(calculate_pos_ratios)
    df = pd.concat([df, ratios], axis=1)


    df.to_excel(f'/content/{label}_pos_ratios.xlsx', index=False)
    print(f'{label} сохранён как {label}_pos_ratios.xlsx')


    medians[label] = ratios.median()


median_df = pd.DataFrame(medians).T
median_df.index.name = 'Group'


median_df.to_excel('/content/pos_ratios_medians.xlsx')
print("Медианы сохранены в pos_ratios_medians.xlsx")
```

auto-semantics index

```python
import pandas as pd
import spacy


nlp = spacy.load("ru_core_news_sm")


def calculate_indices(text):
    doc = nlp(text)
    total_tokens = 0
    content_words = 0
    pronouns = 0

    for token in doc:
        if token.is_alpha:
            total_tokens += 1
            pos = token.pos_
            if pos in ["NOUN", "VERB", "ADJ", "ADV"]:
                content_words += 1
            elif pos == "PRON":
                pronouns += 1

    if total_tokens == 0:
        return 0, 0

    auto_index = content_words / total_tokens
    pron_index = pronouns / total_tokens
    return auto_index, pron_index


df = pd.read_excel("norm.xlsx")
df["auto_semanticity"], df["pronominalization"] = zip(*df["transcript"].map(calculate_indices))


df.to_excel("output_with_indices.xlsx", index=False)
print("Индексы сохранены в output_with_indices.xlsx")
```

proportion of simple sentences

```python
import spacy
import pandas as pd


nlp = spacy.load("ru_core_news_sm")


input_file = "group_sz.xlsx"
df = pd.read_excel(input_file)


def calculate_simple_sentence_ratio(text):
    doc = nlp(text)
    simple_sentences = 0
    total_sentences = 0


    for sent in doc.sents:
        total_sentences += 1
        subject = [token for token in sent if token.dep_ == 'nsubj']
        verb = [token for token in sent if token.pos_ == 'VERB']
        if len(subject) == 1 and len(verb) == 1:
            simple_sentences += 1

    if total_sentences == 0:
        return 0
    return simple_sentences / total_sentences


df['simple_sentence_ratio'] = df['transcript'].apply(calculate_simple_sentence_ratio)

results_df = df[['transcript', 'simple_sentence_ratio']]


median_ratio = results_df['simple_sentence_ratio'].median()


output_file = "results.xlsx"
results_df.to_excel(output_file, index=False)


print(f"Медиана доли простых предложений: {median_ratio}")
print(f"Результаты сохранены в файл {output_file}")
```

parataxis and hypotaxis

```python
import spacy
import pandas as pd


nlp = spacy.load("ru_core_news_sm")


input_file = "group_sz.xlsx"
df = pd.read_excel(input_file)


def calculate_parataxis_hypotaxis(text):
    doc = nlp(text)
    total_sentences = 0
    parataxis_count = 0
    hypotaxis_count = 0

    for sent in doc.sents:
        total_sentences += 1
        if any(tok.pos_ == 'SCONJ' and tok.dep_ in {'mark', 'advcl', 'acl'} for tok in sent):
            hypotaxis_count += 1
        elif any(tok.pos_ == 'CCONJ' and tok.dep_ in {'cc', 'conj'} for tok in sent):
            parataxis_count += 1

    if total_sentences == 0:
        return 0, 0

    return parataxis_count / total_sentences, hypotaxis_count / total_sentences


df[['parataxis_ratio', 'hypotaxis_ratio']] = df['transcript'].apply(
    lambda x: pd.Series(calculate_parataxis_hypotaxis(x))
)


results_df = df[['transcript', 'parataxis_ratio', 'hypotaxis_ratio']]


parataxis_median = results_df['parataxis_ratio'].median()
hypotaxis_median = results_df['hypotaxis_ratio'].median()


output_file = "results_parataxis_hypotaxis.xlsx"
results_df.to_excel(output_file, index=False)


print(f"Медиана доли паратаксиса: {parataxis_median:.3f}")
print(f"Медиана доли гипотаксиса: {hypotaxis_median:.3f}")
print(f"Результаты сохранены в файл {output_file}")
```

```python
import spacy
import pandas as pd


nlp = spacy.load("ru_core_news_sm")


input_file = "group_sz.xlsx"
df = pd.read_excel(input_file)


introductory_words = {
    # Мнение, источник
    "по-моему", "по мнению", "кажется", "как я думаю", "на мой взгляд", "по всей видимости",
    "по слухам", "по рассказам", "как говорят", "по сообщениям", "как сообщают",

    # Неуверенность
    "возможно", "вероятно", "наверное", "по-видимому", "может быть", "пожалуй", "скорее всего", "быть может",

    # Уверенность
    "безусловно", "разумеется", "конечно", "естественно", "несомненно", "определённо", "очевидно", "бесспорно", "действительно",

    # Связь мыслей
    "итак", "таким образом", "значит", "следовательно", "наконец", "впрочем", "между прочим", "во-первых",
    "во-вторых", "во-третьих", "наоборот", "кстати", "вдобавок", "в итоге", "причём",

    # Эмоциональная оценка
    "к счастью", "к несчастью", "к сожалению", "странно", "удивительно", "на редкость", "по правде", "по справедливости",
    "по совести", "по сути", "что интересно", "как ни странно",

    # Формулы общения
    "допустим", "скажем", "видишь ли", "понимаешь", "представь себе", "признайся", "поверь", "будем считать",
    "веришь ли", "помнится", "собственно говоря", "если хотите"
}

def count_introductory_ratio(text):
    doc = nlp(text)
    total_tokens = 0
    intro_count = 0

    for token in doc:
        if token.is_alpha:
            total_tokens += 1
            word_lower = token.text.lower()
            if word_lower in introductory_words:
                intro_count += 1

    if total_tokens == 0:
        return 0
    return intro_count / total_tokens


df['intro_ratio'] = df['transcript'].apply(count_introductory_ratio)


output_file = "results_introductory_words.xlsx"
df.to_excel(output_file, index=False)


median_intro = df['intro_ratio'].median()
print(f"Медианная доля вводных слов/конструкций: {median_intro:.4f}")
print(f"Результаты сохранены в файл {output_file}")
```

proportion of ellipsis

```python
import spacy
import pandas as pd

nlp = spacy.load("ru_core_news_sm")

def is_potential_ellipsis(sent):
    has_verb = any(token.pos_ in {"VERB", "AUX"} and token.dep_ == "ROOT" for token in sent)
    return not has_verb and len(sent) > 1

def get_verb_candidates(sent):
    return [token.lemma_ for token in sent if token.pos_ in {"VERB", "AUX"} and token.dep_ == "ROOT"]

def ellipsis_with_anaphora(text):
    doc = nlp(text)
    sentences = list(doc.sents)
    total = len(sentences)
    ellipsis_count = 0
    prev_verbs = []

    for sent in sentences:
        if is_potential_ellipsis(sent):
            if prev_verbs:
                ellipsis_count += 1
        else:
            prev_verbs = get_verb_candidates(sent)

    if total == 0:
        return 0
    return ellipsis_count / total


df = pd.read_excel("group_sz.xlsx")


df["ellipsis_ratio_contextual"] = df["transcript"].apply(ellipsis_with_anaphora)


df.to_excel("results_ellipsis_contextual.xlsx", index=False)


med = df["ellipsis_ratio_contextual"].median()
print(f"Медианная доля эллипсиса (учитывая контекст): {med:.4f}")
```
